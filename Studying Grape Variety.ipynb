```python
import pandas as pd
import normalizing_text
import numpy as np
from collections import Counter

from wordcloud import WordCloud
from IPython.display import Image
import PIL
import pydot

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn import tree
from sklearn.metrics import classification_report, precision_score, confusion_matrix

import itertools
import gensim
import warnings

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

warnings.filterwarnings('ignore')
pd.options.mode.chained_assignment = None
%matplotlib inline

raw_wine_dataset = pd.read_csv('data/all_scraped_wine_info_chardonnay.csv')
raw_wine_dataset = raw_wine_dataset[['Name', 'Category', 'Variety', 'Appellation', 'Country', 'Province', 'Region', 
                                     'Subregion', 'Winery', 'Designation', 'Vintage', 'Bottle Size', 'Alcohol',
                                     'Description', 'Date Published', 'Importer', 'Price', 'Rating', 'Reviewer',
                                     'Reviewer Twitter Handle', 'User Avg Rating']]
print(raw_wine_dataset.shape)
raw_wine_dataset.head()
# raw_wine_dataset.head(10)
```

```python
sno, descriptor_mapping, punctuation_table, unknowable_tokens = normalizing_text.load_text_normalization_objects(raw_wine_dataset)

wine_description = 'Marmalade, toffee, vanilla and tonka bean swirl on the nose. Rich and smooth, rounded and mellow, yet with an incredibly bright, aromatic core of fine citrus. Think dried and candied orange peel and candied blood orange. What a magical balance the citrus freshness strikes with the cushioning, gentle vanilla. Textural bliss and aromatic fireworks. Drink until 2040, at least.'
normalized_description = normalizing_text.full_text_normalization(wine_description, sno, descriptor_mapping, punctuation_table, unknowable_tokens)
print(normalized_description)
```

```python
raw_wine_dataset = raw_wine_dataset.loc[raw_wine_dataset['Description'].str.len() > 5]
raw_wine_dataset['description_normalized'] = raw_wine_dataset['Description'].apply(lambda x: normalizing_text.full_text_normalization(x, sno, descriptor_mapping, punctuation_table, unknowable_tokens))
raw_wine_dataset[['Name', 'Description', 'description_normalized']]
```

```python
normalized_wine_dataset = raw_wine_dataset[['description_normalized']]
normalized_wine_dataset = normalized_wine_dataset.reset_index()
normalized_wine_dataset = normalized_wine_dataset.description_normalized.apply(pd.Series)\
                                                            .merge(normalized_wine_dataset, left_index=True, right_index=True)\
                                                            .drop(['description_normalized'], axis=1)\
                                                            .melt(id_vars = ['index'], value_name='description_normalized')\
                                                            .drop('variable', axis=1)\
                                                            .dropna()

normalized_wine_dataset.sort_values(['index'], inplace=True)
normalized_wine_dataset.set_index('index', inplace=True)

dummified_descriptors = pd.get_dummies(normalized_wine_dataset['description_normalized'])
final_wine_dataset = dummified_descriptors.groupby(dummified_descriptors.index).sum()

final_wine_dataset.head(10)
print(final_wine_dataset.shape)
```

```python
final_wine_dataset = final_wine_dataset.loc[final_wine_dataset.sum(axis=1) >= 5]

print(final_wine_dataset.shape)
print(final_wine_dataset.info())
```

```python
from sklearn.cluster import KMeans
from kmodes import kmodes

# km = kmodes.KModes(n_clusters=3, init='Huang', n_init=5, verbose=0)
# clusters = km.fit_predict(x)

sum_of_squared_distances = []
K = range(1, 15)
for k in K:
    km = kmodes.KModes(n_clusters=k)
    km.fit(final_wine_dataset)
    sum_of_squared_distances.append(km.cost_)
    print(km.cost_)

plt.plot(K, sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
km6 = KMeans(n_clusters=6)
km6.fit(final_wine_dataset)

final_wine_dataset['cluster_label'] = list(km6.labels_)
just_cluster_labels = final_wine_dataset[['cluster_label']]
```

```python
from collections import Counter

clustered_dataset = pd.merge(left=raw_wine_dataset, right=just_cluster_labels, left_index=True, right_index=True)

plt.figure(figsize=(10, 12))

for n in range(6):
    one_cluster = clustered_dataset.loc[clustered_dataset['cluster_label'] == n]
    one_cluster_text = list(one_cluster['description_normalized'])
    one_cluster_text = [item for sublist in one_cluster_text for item in sublist]

    descriptor_value_counts = Counter(one_cluster_text).most_common(20)
#     print(descriptor_value_counts)
#     print('')

    wordcloud_dict = dict()
    for descriptor in descriptor_value_counts:
        wordcloud_dict[descriptor[0]] = descriptor[1]

    word_cloud = WordCloud(width=100, height=150, background_color='white', relative_scaling=0.60).fit_words(wordcloud_dict)
    
    plt.subplot(2, 3, n+1)   

    plt.imshow(word_cloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(n, fontsize=13)

plt.subplots_adjust(wspace=0.1, hspace=0)
plt.suptitle('Most Common Descriptors by Grape Variety', fontsize=20)
plt.show()
# clustered_dataset.head()
```
